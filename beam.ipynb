```python

import pandas as pd
import apache_beam as beam
import apache_beam.runners.interactive.interactive_beam as ib
from apache_beam.runners.interactive.interactive_runner import InteractiveRunner

def SplitRow(row):
  return row.split(',')

def ApplyFilterByFinance(row):
  return row[3] == 'Finance'

p1 = beam.Pipeline(InteractiveRunner())

collection = (p1
    | 'Read File' >> beam.io.ReadFromText('/content/dept_data.txt'))

finance = (collection
    | 'Split Rows' >> beam.Map(SplitRow)
    | 'Filter' >> beam.Filter(ApplyFilterByFinance)
    | 'Prepare' >> beam.Map(lambda row: (row[1], 1))
    | 'Sum' >> beam.CombinePerKey(sum)
)

p1.run()

ib.collect(finance)
```
